# Seasonal decomposition and checking stationarity

For the time series modeling of vehicle counts we'll be using various models to predict the vehicle count forecast. Most of these models work by breaking down the observation into a trend component and a seasonal component.

As such, we will observe these trend and seasonal component by conducting a seasonal decomposition.

```python
# using statsmodel's seasonal decompose function
decomp_result = seasonal_decompose(traffic_data['total'].dropna(), model='add', period=7*48)

plot_observed = decomp_result.observed
plot_trend = decomp_result.trend
plot_seasonal = decomp_result.seasonal
plot_residual = decomp_result.resid
```

```python
fig, ax = plt.subplots(4, 1, figsize=(20, 12), sharex=True)

decomp_result.observed.plot(ax=ax[0], ylabel='Observation')
decomp_result.trend.plot(ax=ax[1], ylabel='Trend')
decomp_result.seasonal.plot(ax=ax[2], ylabel='Seasonal')
decomp_result.resid.plot(ax=ax[3], ylabel='Residual')

# plotting black line demarkating every week
for date in traffic_data.resample('D').mean()[traffic_data.resample('D').mean()['day_of_week']==0].index:
    for axid in ax:
        axid.axvline(date, c='w', lw=1.5)
        
fmt_plt(plt, plot_legends=False, ylabel='Residual')
plt.xlim(dt(2022,10,3),dt(2022,11,14))
```

The figure above shows the seasonal decomposition of vehicle count time series. The following observations can be made:
- There are clear weekly seasonality in the observation
- The trend component it's relatively minor compare to the seasonal component (+/- 0.5 for trend vs +/- 5)
- The traffic count data is fairly noisy, as indicated by the high residual components

In order to assess if the model is stationary, i.e.: the trend component is negligible, we shall conduct an ADF test.

```python
def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(traffic_data['total'])
```

We can see that ADF test results shows that the traffic account time series is stationary.

## Feature scaling

Some time series model requires the features to be scaled in order to facilitate faster model convergence. As such, we will scale the vehicle count feature here using the Darts scaler.

The Darts scaler used in the study, is the default Darts scaler which is a MinMax scaler. Only the training data is scaled here to prevent data leakage. After making predictions, the same scaler can then be used to inverse transform the predictions to the original scale.

```python
scaler = Scaler()
train_sc = scaler.fit_transform(train)
```

# Function definition for plotting and evaluating forecasts

```python
def predict(model, n_forecast, model_name='', display_n_before=48*0, plot=True, scaling=True, probalistic=False):
    '''
    This function takes a model and a feature, and generates the prediction for the next n_forecast values
    '''

    # getting the predictions using the model
    if probalistic: # define the number of samples for probabilistic predictions
        preds = model.predict(n_forecast, num_samples=500)
    else:
        preds = model.predict(n_forecast)

    # scaling the predictions if using scaled inputs
    if scaling:
        preds = scaler.inverse_transform(preds)

    # calculate model RMSE
    model_rmse = rmse(actual_series=[traffic_data_ts], pred_series=[preds])

    # plot if requested
    if plot:
        plt.figure(figsize=(7, 5))

        # split point between train and validation
        split = train_sc.end_time()

        # slicing the series before and after the split point
        after = split - pd.Timedelta(minutes=30)*display_n_before
        before = split + pd.Timedelta(minutes=30)*n_forecast

        traffic_data_ts.slice(after, before).plot(label="actual")
        preds.plot(label="forecast")

        plt.title(
            f'{model_name} | RMSE={model_rmse:.3f}')
        
        fmt_plt(plt,plot_legends=False)
        
    # if not for plotting return model_mase and model_rmse
    else:
        return model_rmse

    
    
def export_result(model, model_name, n_forecast, scaling=True):
    '''
    This function takes a model and a feature and calculates the RMSE for a set of pre-determined forecasting horizons.
    Returns a dataframe of RMSE values.
    '''
    
    # instantiate a dictionary to contain the model name and RMSE score
    score_dict = {'model':model_name,
                  'RMSE': [],
                 }

    # calculate the model RMSE using the previous function
    if scaling: 
        model_rmse = predict(model, n_forecast, plot=False)
    else:
        model_rmse = predict(model, n_forecast, plot=False, scaling=False)

    # appending the model RMSE to the dictionary
    score_dict['RMSE'].append(model_rmse)
    
    # converting reduction rate to a data frame row and returning it
    return pd.DataFrame(score_dict)
```

# Modelling and Predictions

In this section we will be comparing various models in order to generate prediction / forecasts for future traffic counts.

As previously mentioned, we will be using the previous 5 weeks of data to train the model and use the model to predict the next one week of traffic.

```python
prediction_duration = 1*7*48 # predict the next 1 week (1 week x 7 days x 48 half-hour periods per day)
seasonality = 7*48
```

## Model 1 - Naive Mean (baseline)

This model always predicts the mean of the training dataset.

```python
# instantiate the model
mean = NaiveMean()
mean.fit(train_sc)

# plotting the forecast
predict(model=mean, model_name='Naive Mean (baseline)', n_forecast=prediction_duration)

# storing scores for comparison
mean_scores = export_result(model=mean, model_name='Naive Mean', n_forecast=prediction_duration)
```

## Model 2 - Naive Seasonal

This model always predicts the value of K time steps ago, where K is the seasonality duration. 

When K=1, this model predicts the last value of the training set. When K>1, it repeats the last K values of the training set.

Based on the time series decomposition, the value of K should be set to 48\* 7 (48 half hour intervals per day \* 7 days per week)

```python
# instantiate the model
naive = NaiveSeasonal(K=seasonality)
naive.fit(train_sc)

# plotting the forecast
predict(model=naive, model_name='Naive Seasonal', n_forecast=prediction_duration)

# storing scores for comparison
naive_scores = export_result(model=naive, model_name='Naive Seasonal', n_forecast=prediction_duration)
```

## Model 3 - Holt Winters Exponential Smoothing

Forecasts produced using exponential smoothing methods are weighted averages of past observations, with the weights decaying exponentially as the observations get older. [[source]](https://otexts.com/fpp2/expsmooth.html)

The Holt-Winters seasonal method comprises of the forecast equation and three smoothing equations  level, trend, and seasonal components.

```python
# instantiate the model
hw = ExponentialSmoothing(trend=ModelMode.NONE, seasonal=ModelMode.ADDITIVE, seasonal_periods=seasonality)
hw.fit(train_sc)

# plotting the forecast
predict(model=hw, model_name="Holt Winter's", n_forecast=prediction_duration)

# storing scores for comparison
hw_scores = export_result(model=hw, model_name="Holt Winter's", n_forecast=prediction_duration)
```

## Model 4 - Theta Model

The Theta model of Assimakopoulos & Nikolopoulos (2000) is a simple method for forecasting the involves fitting two-lines, forecasting the lines using a Simple Exponential Smoother, and then combining the forecasts from the two lines to produce the final forecast. [[source]](https://www.statsmodels.org/devel/examples/notebooks/generated/theta-model.html)

```python
# instantiate the model
theta = Theta(theta=2, seasonality_period=7*48, season_mode=SeasonalityMode.ADDITIVE)
# using the unscaled data for theta
theta.fit(train)

# plotting the forecast
predict(model=theta, model_name="Theta", n_forecast=prediction_duration, scaling=False)

# storing scores for comparison
theta_scores = export_result(model=theta, model_name="Theta", n_forecast=prediction_duration, scaling=False)
```

## Model 5 - FB Prophet

Prophet procedure is an additive regression model with four main components [[source]](https://research.facebook.com/blog/2017/2/prophet-forecasting-at-scale/):
- A piecewise linear or logistic growth curve trend. Prophet automatically detects changes in trends by selecting changepoints from the data.
- A yearly seasonal component modeled using Fourier series.
- A weekly seasonal component using dummy variables.
- A user-provided list of important holidays.

Additionally the Prophet model also allows users to add their own seasonality components

```python
# instantiate the model
prophet = Prophet(add_seasonalities=[{
                                                'name': 'weekly-7*48',  # (name of the seasonality component),
                                                'seasonal_periods': 7*48,  # (nr of steps composing a season),
                                                'fourier_order': 100,  # (number of Fourier components to use),
                                                'mode': 'additive',  # ('additive' or 'multiplicative')
                                                },
                                                {
                                                'name': 'daily-1*48',  # (name of the seasonality component),
                                                'seasonal_periods': 1*48,  # (nr of steps composing a season),
                                                'fourier_order': 100,  # (number of Fourier components to use),
                                                'mode': 'additive',  # ('additive' or 'multiplicative')
                                                }],
                             )

# using the unscaled data for prophet
prophet.fit(train)

# plotting the forecast
predict(model=prophet, model_name="Prophet", n_forecast=prediction_duration, scaling=False)

# storing scores for comparison
prophet_scores = export_result(model=prophet, model_name="Prophet", n_forecast=prediction_duration, scaling=False)
```

# Effect of Forecast Horizon

When deploying the model into production, we have the choice of the frequency in which we will be generating the model (e.g.: generate weekly prediction for the next week, or generate daily prediction for the next day)

In this section, we will take a look at the effect of different forecast horizon on the accuracy of the model.

To do this, we will conduct a backtest using the prophet model, where the model is repeatedly trained and tested according to the forecast horizon. The backtesting period used is between 17 October to 15 November.

```python
weekly_train_scores = prophet.backtest(series=traffic_data_ts, 
                                       start=pd.Timestamp(year=2022, month=10, day=17), 
                                       forecast_horizon=7*48, stride=7*48, # getting weekly forecasts
                                       metric=rmse, reduction=None, verbose=True,
                                      )

threedays_train_scores = prophet.backtest(series=traffic_data_ts, 
                                      start=pd.Timestamp(year=2022, month=10, day=17), 
                                      forecast_horizon=3*48, stride=3*48, # getting daily forecasts
                                      metric=rmse, reduction=None, verbose=True,
                                     )

daily_train_scores = prophet.backtest(series=traffic_data_ts, 
                                      start=pd.Timestamp(year=2022, month=10, day=17), 
                                      forecast_horizon=1*48, stride=1*48, # getting daily forecasts
                                      metric=rmse, reduction=None, verbose=True,
                                     )
```

```python
rmse_scores_backtest = pd.concat([pd.DataFrame(weekly_train_scores,columns=['7 days']).unstack().reset_index(level=[0]),
                                  pd.DataFrame(threedays_train_scores,columns=['3 days']).unstack().reset_index(level=[0]),
                                  pd.DataFrame(daily_train_scores,columns=['1 days']).unstack().reset_index(level=[0]),
                                 ])

rmse_scores_backtest.columns = ['Forecast Horizon','RMSE']

rmse_scores_backtest_mean = rmse_scores_backtest.groupby('Forecast Horizon').mean()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(rmse_scores_backtest,
             x='RMSE',
             hue='Forecast Horizon',
             kde=True,
             legend=False
            )

labels = [f'{horizon} | RMSE = {RMSE:.2f}' for horizon, RMSE in rmse_scores_backtest_mean.to_dict()['RMSE'].items()]
plt.legend(title='Forecast Horizon', loc='upper right', labels=labels, frameon=True)
```

Only about 6% difference in prediction accuracy between 7 days and 1 day, hence we will use a forecast horizon of 7 days.
